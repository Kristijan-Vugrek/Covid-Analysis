```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import requests
import pandas as pd 
import io
import sys
from pandas import DataFrame
```

```python
sys.stdout = io.TextIOWrapper(io.BytesIO(), encoding=sys.stdout.encoding)
driver = webdriver.Chrome(
    executable_path='C:\Program Files (x86)\chromedriver.exe')
wait = WebDriverWait(driver, 3)

start_time = time.time()
data = []
df = pd.DataFrame(data)

driver.get('https://www.electromaps.com/en/charging-stations/angola')

driver.maximize_window()
time.sleep(1)
    
cookie = driver.find_element(
    By.XPATH, '//button[@id="CybotCookiebotDialogBodyLevelButtonLevelOptinAllowAll"]')
try:
    cookie.click()
except:
    pass


def load_more_stations():
    while True:
        try:
            load_more = wait.until(EC.element_to_be_clickable(
                (By.XPATH, '//a[@id="loadMoreStations"]')))
            load_more.click()
            time.sleep(0.5)
        except:
            break


def load_more():
    while True:
        try:
            load_more = driver.find_element(By.XPATH, '//a[@id="loadMore"]')
            load_more.click()
            time.sleep(0.1)
        except:
            break


load_more()

wait.until(EC.element_to_be_clickable(
            (By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')))
locations_A = driver.find_elements(
    By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')
driver.implicitly_wait(1)

for location_A in locations_A:
    driver.execute_script("arguments[0].click();", location_A)
    time.sleep(1)
    load_more()
    

    locations_B = driver.find_elements(
        By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')

    for location_B in locations_B:
        driver.execute_script("arguments[0].click();", location_B)
        time.sleep(1)
        load_more_stations()

        wait.until(EC.element_to_be_clickable(
            (By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')))
        locations_C = driver.find_elements(
            By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')

        for location_C in locations_C:
            driver.execute_script("arguments[0].click();", location_C)
            time.sleep(2)
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            div_tag_addresses = soup.find('div', {'class' : 'char-aside-info-text'})
            div_tag_names = soup.find('div', {'class' : 'char-heading-title'})
            div_tag_connectors = soup.find_all('div', {'class' : 'char-connectors-item'})
            div_tag_location = soup.find('div', {'class' : 'breadcrumbs desktop'})

            for connector in div_tag_connectors:
                connector_amount = connector.find('span').text
                connector_type = connector.find('h3', {'class': 'xs'}).text
                connector_power = connector.find('div', {'class' :'char-kw'}).find('span').text
                    
                if not connector_amount:
                        connector_amount = ''
                if not connector_type:
                        connector_type = ''
                if not connector_power:
                        connector_power = ''

                text_locations = div_tag_location.find_all('a', {'class': 'breadcrumb-item w-inline-block'})
                text_location_2 = text_locations[2].find('div', {'class': 'body-s'}).text
                text_location_3 = text_locations[3].find('div', {'class': 'body-s'}).text
                text_address = div_tag_addresses.find('div', {'class': 'body-s'}).text
                text_name = div_tag_names.find('span', {'class': 'title-charge-point'}).text
                data.append({'name': text_name, 'address': text_address, 'amount': connector_amount, 'type': connector_type, 'power': connector_power, 'regija' : text_location_2, 'grad' : text_location_3})
                df
            

            driver.execute_script("window.history.go(-1)")
            wait.until(EC.element_to_be_clickable(
                (By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')))
        locations_C = driver.find_elements(
                By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')
        if location_C == locations_C[-1]:
                pass

        driver.execute_script("window.history.go(-1)")
        wait.until(EC.element_to_be_clickable(
            (By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')))
    locations_B = driver.find_elements(
            By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')
    if location_B == locations_B[-1]:
            pass

    driver.execute_script("window.history.go(-1)")
    wait.until(EC.element_to_be_clickable(
        (By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')))
locations_A = driver.find_elements(
        By.XPATH, '//a[@class="chargepoints-stations-item w-inline-block show"]')
if location_A == locations_A[-1]:
        if len(data) > 0:
         df.to_excel('punionice.xlsx', index=False)
        driver.quit
df
```
